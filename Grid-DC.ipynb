```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['bgr', 'bu', 'sc', 'pcv', 'wc']]
dependent = dataset['classification_yes']
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import pickle
scaler_filename = 'sc.pkl'
with open(scaler_filename, 'wb') as file:
    pickle.dump(sc, file)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini','entropy', 'log_loss'],
              'max_features': ['sqrt','log2'],
              'splitter':['best','random'],
             'class_weight':['balanced']}  

grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit = True, verbose = 3, n_jobs = -1, scoring = 'f1_weighted')

grid.fit(X_train, y_train) 
```

```python
# print best parameter after tuning
re = grid.cv_results_
grid_predictions = grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, grid_predictions, average = 'weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_), f1_macro)
```

```python
print("The confusion Matrix:\n", cm)
```

```python
print("The report:\n", clf_report)
```

```python
from sklearn.metrics import roc_auc_score 
#Receiver Operating Characteristic_Area Under Curve
#if the curve is L shape then the classification is 100% correct

roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
# (: = rows) & (1 = Column)
#grid.predict_proba(X_test)->inbuilt function(probability compulsory)
```
